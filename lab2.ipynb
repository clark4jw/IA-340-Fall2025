```python
df = df[(df['X']>=0) & (df['Y']>=0) & (df['CCN']>=0)] #filter data
df.describe()
```

I selected CCN to be display, I sorted the data by CCN number, and filtered the data by ensuring the X, Y, and CCN columns were above 0

```python
avg_coords = df.groupby("CCN").agg(
    Avg_X=("X", "mean"),
    Avg_Y=("Y", "mean")
).reset_index()

print(avg_coords)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(avg_coords["Avg_X"], avg_coords["Avg_Y"], color="blue", alpha=0.7)

# Add labels
plt.title("Where in D.C. is there Crime?")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
```

This grouped data shows on a scatterplot a visual representation of crime hotspots in Washington, D.C. This visualization displays the X and Y coordinates of each report. Also, it only includes where crime has occurred, outlining the city. 

```python
plt.hexbin(df["X"], df["Y"], gridsize=30, cmap="Blues")
plt.colorbar(label="Case Count")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Density of Cases")
plt.show()
```

This visualization shows the X and Y grouped data in a heatmap of D.C. On this visualization you are able to see which grouped hotspots have the most crime occurring. As the color gets darker, more incidents occurred.
